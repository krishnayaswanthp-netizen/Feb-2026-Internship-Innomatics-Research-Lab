Problem 1: Unique Words in a Sentence


```python
sentence="python is easy and python is powerful"
l=sentence.split()
s=set(l)
print(f"the unique words are: {s} and the length is :{len(s)}")
```

Problem 2: Highest Salary from Employee Data


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_key = max(employees,key=employees.get)
print(f"The highest salary is {employees[highest_key]}-{highest_key}")
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
threshold=1000
c=0
for i in prices:
  if i > 1000:
    c+=1
print("the products above the threshold are: ",c)
```
