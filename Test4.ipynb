Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid


```python
def recharge(n):
  l=[199,299,399,599]
  while True:
    if n in l:
      print("Recharge Successful")
      break
    else:
      print("Recharge Failed")
n=int(input("enter the recharge you try to provide: "))
recharge(n)
```

Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15

Print Stock OK otherwise


```python
d={'laptop':10,
   'mobiles':20,
   'toys':30,
   'tube lights':40,
   'charis':50}

for i,j in d.items():
  if j<15:
    print(i," - Reorder Alert")
  else:
    print(i," - Stock OK")
```

Problem 3: Student Result Processing System
Problem Statement

Create a function that:
Accepts a list of student marks

Calculates average marks

Prints Pass if average ≥ 50 else Fail

```python

def avg_mark(l):
  n=len(l)
  avg=sum(l)/n

  if avg>=50:
    return 'pass'
  else:
    return 'fail'

l=list(map(int,input().split()))
print(avg_mark(l))
```

Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50

₹12 per km

Peak hour → 25% extra

Allow retry using while loop


```python
def cal(dis,peak=4):
  base_fare=50
  fare=base_fare+12*dis
  peak1=8
  peak2=6

  if peak==peak1 or peak==peak2:
    fare=fare+(fare*0.25)
  return fare
dis=int(input("enter the distance: "))
peak=int(input("enter the peak time: "))
print(cal(dis,peak))
```

Problem 5: Employee Attendance Eligibility Checker Problem Statement

Create a function that:
Accepts attendance list ("P", "A")

Returns Eligible if ≥ 75% else Not Eligible

```python
l=['P','P','P','P','P','A','A','P']

n=len(l)
c=l.count('P')

attendence=(c/n)*100

if attendence>=75:
  print("eligible")
else:
  print("not eligible")
```

Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


> Add blockquote


Must contain at least one digit

Must contain at least one special character (@#$)


```python
def password(s):
  n=len(s)
  has_digit=False
  has_special=False

  for i in s:
    if i.isdigit():
      has_digit=True
      break
  for i in s:
    if i in ['@','#','$']:
      has_special=True
      break
  if n<8:
    return 'the password is weak use more than 8 characters'
  elif not has_digit:
    return 'the password is weak use atleast one digit'
  elif not has_special:
    return 'the password is weak use atleast one special character'
  else:
    return 'the password is strong'
print(password(input("enter the password you thought of: ")))
```
