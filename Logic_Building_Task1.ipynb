1. User Login Check:

Problem Statement:
Given a username and password, check whether login is successful.

username = "admin"

password = "1234"


```python
username = "admin"
password = "1234"
s1=input("user name :")
s2=input("password :")
if s1==username and s2==password:
  print("login successful")
else:
  print("login failed")
```

2. Pass / Fail Analyzer

Problem Statement:
Analyze student results.

marks = [45, 78, 90, 33, 60]



```python
count_of_passed=0
count_of_failed=0
marks = [45, 78, 90, 33, 60]
for i in marks:
  if i<=50:
    count_of_failed+=1
  else:
    count_of_passed+=1
print("count of passed students : ",count_of_passed)
print("count of failed students : ",count_of_failed)
```

3. Simple Data Cleaner

Problem Statement:
Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]


```python
names = [" Alice ", "bob", " CHARLIE "]
k=[]
for i in names:
  k.append(i.strip().lower())
print(k)
```

4. Message Length Analyzer

Problem Statement:
Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]



```python
messages = ["Hi", "Welcome to the platform", "OK"]

for i in messages:
  if len(i)>10:
    print(i," : ",len(i)," (flaged)")
  else:
    print(i," : ",len(i))
```

5. Error Message Detector

Problem Statement:
Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]



```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
c=0
for i in range(len(logs)-1):
  if logs[i]=="ERROR":
    c+=1
    logs.pop(i)
print(logs)
print(c)
```
