Problem Statement 1:Smart Parking Lot Management System

```python
Parking_Capacity= 50
original= Parking_Capacity
Vehicle_Logs= ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
for i in Vehicle_Logs:
  if i=='IN':
    Parking_Capacity-=1
  elif i=='OUT':
    Parking_Capacity+=1
print(f'Currently Parked Vehicles: {original-Parking_Capacity}')
if Parking_Capacity>0:
  print(f'Parking Capacity: Available')
else:
  print('Sorry, Parking Lot is Full')
```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic_level, weather):
    base_time = distance * 5

    if traffic_level == "High":
        base_time += 10
    elif traffic_level == "Medium":
        base_time += 5

    if weather == "Rainy":
        base_time += 5
    elif weather == "Stormy":
        base_time += 15

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = 0

    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)


booked = [1] * 150
analyze_occupancy(200, booked)
```

Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0

    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", average, "%")
    print("Server Status:", status)


classify_server_load([45, 60, 70, 85, 90])
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resources(resource_usage):
    overused = []

    for resource in resource_usage:
        if resource_usage[resource] > 8:
            overused.append(resource)

    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", overused)
    print("Energy Alert:", alert)


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

 Problem 6: Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, total_registrations):
    if total_registrations <= capacity:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


manage_registrations(100, 105)
```
