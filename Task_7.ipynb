```python
class Order:
  status_flow=['placed','confirmed','out for order','delivered']
  def __init__(self,order_id,restaurant):
    self.order_id=order_id
    self.restaurant=restaurant
    self.current_step=0
  def update_status(self):
    if self.current_step==len(self.status_flow)-1:
      print('order is delivered')
      return
    self.current_step+=1
    print(f'order is {self.status_flow[self.current_step]}')
    return
  def show_summary(self):
    print(f'order id is {self.order_id}')
    print(f'restaurant is {self.restaurant}')
    print(f'status: {self.status_flow[self.current_step]}')
order1=Order('101','pizza hut')
order1.show_summary()
order1.update_status()
order1.update_status()
order1.update_status()
```

Problem 1: Smart Light Controller
Problem Statement

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class bulb:
  def __init__(self):
    self.status='off'
  def turn_on(self):
    if self.status=='on':
      print('bulb is already on')
      return
    self.status='on'
    print('bulb is on')
    return
  def turn_off(self):
    if self.status=='off':
      print('bulb is already off')
      return
    self.status='off'
    print('bulb is off')
    return
bulb1=bulb()
bulb1.turn_on()
bulb1.turn_on()
bulb1.turn_off()
bulb1.turn_off()
```

Problem 3: Mobile Contact Record
Problem Statement

Create a class to store contact name and phone number and display contact information.


```python
class contact:
  contact=[]
  def __init__(self,name,phone_number):
    self.name=name
    self.phone_number=phone_number
  def add_contact(self):
    self.contact.append((self.name,self.phone_number))
  def display_contact(self):
    for name,phone_number in self.contact:
      print(f'Contact Save')
      print(f'name: {name}')
      print(f'phone number: {phone_number}')
Contact_Name= 'Anita'
Phone_Number= '9876543210'

c1=contact(Contact_Name,Phone_Number)
c1.add_contact()
c1.display_contact()
```
